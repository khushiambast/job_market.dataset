```python
data["Min Salary"] = pd.to_numeric(data["Min Salary"], errors="coerce")
data["Max Salary"] = pd.to_numeric(data["Max Salary"], errors="coerce")
data["Min Experience"] = pd.to_numeric(data["Min Experience"], errors="coerce")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
for column in ["Company Size", "Salary"]:
    data[column].fillna(data[column].mode()[0], inplace=True)
for column in ["Min Salary", "Max Salary", "Min Experience"]:
    data[column].fillna(data[column].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
target = "Min Experience"
categorical_features = ["Role", "Companies", "Location", "Experience", "Skills", 
                        "Company Size", "Eligibility Criteria", "Enrollment Type", "Salary"]
numerical_features = ["Min Salary", "Max Salary"]
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
encoder = OneHotEncoder(drop="first", sparse_output=False)
encoded_features = encoder.fit_transform(data[categorical_features])
```

```python
X = np.hstack((encoded_features, data[numerical_features].values))
```

```python
# Standardize numerical features (for those columns that are numerical)
scaler = StandardScaler()
X[:, -len(numerical_features):] = scaler.fit_transform(X[:, -len(numerical_features):])
```

```python
print("Shape of X (combined features):", X.shape)
```

```python
Y = data[target]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=20)
```

```python
#Linear Regression (for continuous target variable)
from sklearn.linear_model import LinearRegression
logistic_model = LinearRegression()
logistic_model.fit(X_train, Y_train)
Y_pred = logistic_model.predict(X_test)
```

```python
logistic_accuracy = logistic_model.score(X_test, Y_test)
logistic_accuracy
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=20)
kmeans.fit(X)
```

```python
from sklearn.metrics import silhouette_score
kmeans_silhouette = silhouette_score(X, kmeans.labels_)
kmeans_silhouette
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, Y_pred, color='blue', alpha=0.5)
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='red', linestyle='--')
plt.title("Actual vs Predicted: Min Experience")
plt.xlabel("Actual Min Experience")
plt.ylabel("Predicted Min Experience")
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) 
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.7)
plt.title("K-Means Clustering (PCA Reduced)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster')
plt.show()
```

```python
residuals = Y_test - Y_pred
plt.figure(figsize=(10, 6))
plt.scatter(Y_pred, residuals, alpha=0.5, color="purple")
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Predicted Min Experience")
plt.ylabel("Residuals")
plt.title("Residual Plot for Linear Regression")
plt.show()
```

```python
range_n_clusters = [2, 3, 4, 5]
sil_scores = [] 
```

```python
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=20)
    kmeans.fit(X)
    sil_score = silhouette_score(X, kmeans.labels_)
    sil_scores.append(sil_score)

plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, sil_scores, marker='o', linestyle='-', color='b')
plt.title("Silhouette Score for K-Means Clustering")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```
